```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from mpl_toolkits.mplot3d import Axes3D

X=np.array([[89,91,95],[33,20,98],[95,93,88],[40,45,56],[80,86,92],[20,55,12],[52,60,80],[90,95,20],[99,89,60],[85,82,75]])
Y=np.array([1,0,1,0,1,0,0,0,0,1])

X = X[np.logical_or(Y==0,Y==1)]
Y = Y[np.logical_or(Y==0,Y==1)]

model = svm.SVC(kernel='linear')
clf = model.fit(X, Y)

z = lambda x,y: (-clf.intercept_[0]-clf.coef_[0][0]*x -clf.coef_[0][1]*y) / clf.coef_[0][2]

tmp = np.linspace(0,100,110)
x,y = np.meshgrid(tmp,tmp)

fig = plt.figure(figsize=(8,8))
ax  = fig.add_subplot(111, projection='3d')
ax.plot3D(X[Y==0,0], X[Y==0,1], X[Y==0,2],'v',color='green',markersize=6)
ax.plot3D(X[Y==1,0], X[Y==1,1], X[Y==1,2],'o',color='red',markersize=6)
ax.plot_surface(x, y, z(x,y))
ax.set_xlabel('Feature1')
ax.set_ylabel('Feature2')
ax.set_zlabel('Feature3')
ax.view_init(15, 2700)
plt.show()
```
